```python
import pandas as pd
import numpy as np
from autoxgb import AutoXGB
```

```python
!tree ../input/tabular-playground-series-nov-2021/
```

```python
sample_sub = pd.read_csv('../input/tabular-playground-series-nov-2021/sample_submission.csv')
```

```python
axgb = AutoXGB(
    train_filename="../input/tabular-playground-series-nov-2021/train.csv",
    output="output_1",
    test_filename="../input/tabular-playground-series-nov-2021/test.csv",
    targets=['target'],
    use_gpu=True,
    task = "classification",
    idx = None,
    features = None,
    categorical_features = None,
    seed=42,
    num_trials=1000,
    time_limit=25000,
    fast=False,
)
axgb.train()
```

```python
!tree /kaggle/working/output_1
```

```python
sample_output = pd.read_csv('/kaggle/working/output_1/test_predictions.csv')
sample_output.head(10)
```

```python
sample_output = sample_output.drop(['0'], axis=1)
sample_output.head(10)
```

```python
sample_output = sample_output.rename(columns={'1':'target'})
sample_output.head(10)
```

```python
sample_output.to_csv('/kaggle/working/output_1/submission.csv', index=False)
```
